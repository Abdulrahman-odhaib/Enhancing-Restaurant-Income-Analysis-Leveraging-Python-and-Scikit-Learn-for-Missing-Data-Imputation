## So, i follow these steps to involve imputation strategy

### 1** Import Libraries:

#### Import the pandas library as pd for data manipulation and the SimpleImputer class from scikit-learn for imputing missing values.

### 2** Load Dataset:

#### Load the dataset from the Excel file named 'your_dataset.xlsx' into a pandas DataFrame using the pd.read_excel() function.

### 3** Identify Columns with Missing Values:

#### Define a list unlabeled_columns_with_missing_values containing the column names ('Column1', 'Column2', 'Column3') where missing values need to be imputed.

### 4** Impute Missing Values:

#### Create a SimpleImputer object imputer with the strategy set to 'mean' to impute missing values with the mean of each column.
#### Apply the imputation strategy to the specified columns using the fit_transform() method of the SimpleImputer object.
#### Update the DataFrame with the imputed values.

### 5** Check Missing Values Before Imputation:

#### Print the number of missing values in the specified columns before the imputation process using the isnull().sum() method.

### 6** Impute Missing Values Again:

#### Reapply the imputation strategy to the specified columns to ensure consistency.

### 7** Check Missing Values After Imputation:

#### Print the number of missing values in the specified columns after the imputation process to verify that all missing values have been imputed successfully.

### 8** Inspect Data:

#### Print the first few rows of the DataFrame to inspect if the missing values have been replaced correctly.

### 9** Write Updated DataFrame to Excel File:

#### Write the updated DataFrame back to an Excel file named 'Kitap1_updated.xlsx' using the to_excel() method. Setting index=False ensures that the DataFrame index is not included in the Excel file.

### Here is the code i used to done these steps in python

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Load your dataset
data = pd.read_excel('your_dataset.xlsx')

# Identify unlabeled fields (columns) with missing values
unlabeled_columns_with_missing_values = ['Column1', 'Column2', 'Column3']

# Impute missing values using mean strategy
imputer = SimpleImputer(strategy='mean')
data[unlabeled_columns_with_missing_values] = imputer.fit_transform(data[unlabeled_columns_with_missing_values])

# Check for missing values before imputation
print("Missing values before imputation:")
print(data[unlabeled_columns_with_missing_values].isnull().sum())

# Impute missing values using mean strategy
imputer = SimpleImputer(strategy='mean')
data[unlabeled_columns_with_missing_values] = imputer.fit_transform(data[unlabeled_columns_with_missing_values])

# Check for missing values after imputation
print("Missing values after imputation:")
print(data[unlabeled_columns_with_missing_values].isnull().sum())

# Check for missing values after imputation
print("Missing values after imputation:")
print(data[unlabeled_columns_with_missing_values].isnull().sum())

# Print the first few rows to inspect if the missing values are replaced
print(data.head())

# Write the updated DataFrame back to the Excel file
data.to_excel('your_updated_dataset.xlsx', index=False)
```
